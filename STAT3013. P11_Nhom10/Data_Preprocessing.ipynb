```python
def rename_and_duplicate_column(df):
    """
    Renames the 'Price' column to 'Close' and creates another column
    named 'Adj. Close' which is a copy of the 'Close' column.
    """
    # Rename the 'Price' column to 'Close'
    df.rename(columns={'Price': 'Close'}, inplace=True)
    df.rename(columns={'Vol.': 'Volume'}, inplace=True)
    
    # Create a new column 'Adj. Close' as a copy of the 'Close' column
    df['Adj. Close'] = df['Close']
    
    return df
```

```python
import pandas as pd

def fill_missing_dates(df, date_column):
    """
    Fill missing dates in a DataFrame and forward/backward fill missing values.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        date_column (str): The name of the date column.

    Returns:
        pd.DataFrame: A DataFrame with missing dates filled and forward/backward filling applied.
    """
    # Convert date column to datetime
    df[date_column] = pd.to_datetime(df[date_column])
    
    # Create a complete date range
    full_date_range = pd.date_range(start=df[date_column].min(), end=df[date_column].max(), freq='D')
    
    # Reindex the DataFrame with the full date range
    df = df.set_index(date_column).reindex(full_date_range).rename_axis(date_column).reset_index()
    
    # Apply forward fill and backward fill
    df.fillna(method='ffill', inplace=True)  # Forward fill
    df.fillna(method='bfill', inplace=True)  # Backward fill (to handle leading NaN)
    
    return df
```

```python
# Assuming `Gold` is your DataFrame and 'Date' is the date column
Gold_filled = fill_missing_dates(Gold, 'Date')

# Display the updated DataFrame
Gold_filled
```

```python
import pandas as pd

def remove_time_from_date(df, date_column):
    """
    Removes the timezone and time component from a date column in a DataFrame.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        date_column (str): The name of the column containing the date and time.

    Returns:
        pd.DataFrame: The updated DataFrame with the timezone and time removed from the date column.
    """
    # Convert to datetime while handling timezone
    df[date_column] = pd.to_datetime(df[date_column], utc=True)
    
    # Remove timezone and keep only the date part (YYYY-MM-DD)
    df[date_column] = df[date_column].dt.tz_localize(None).dt.date
    
    return df
```

```python
# Apply the function to your DataFrame
Imus = remove_time_from_date(Imus, 'Date')

# Display the updated DataFrame
print(Imus.head())
```
