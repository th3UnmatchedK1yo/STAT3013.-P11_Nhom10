```python
import pandas as pd

def fill_missing_dates(df, date_column):
    """
    Fill missing dates in a DataFrame and forward/backward fill missing values.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        date_column (str): The name of the date column.

    Returns:
        pd.DataFrame: A DataFrame with missing dates filled and forward/backward filling applied.
    """
    # Convert date column to datetime
    df[date_column] = pd.to_datetime(df[date_column])
    
    # Create a complete date range
    full_date_range = pd.date_range(start=df[date_column].min(), end=df[date_column].max(), freq='D')
    
    # Reindex the DataFrame with the full date range
    df = df.set_index(date_column).reindex(full_date_range).rename_axis(date_column).reset_index()
    
    # Apply forward fill and backward fill
    df.fillna(method='ffill', inplace=True)  # Forward fill
    df.fillna(method='bfill', inplace=True)  # Backward fill (to handle leading NaN)
    
    return df
```

```python
import pandas as pd

def clean_commas(df, columns):
    """
    Removes commas from specified columns in a DataFrame.
    
    Args:
    df (pd.DataFrame): The input DataFrame.
    columns (list): List of column names to clean commas from.
    
    Returns:
    pd.DataFrame: DataFrame with commas removed in specified columns.
    """
    for column in columns:
        if column in df.columns:
            # Remove commas and convert to numeric type if necessary
            df[column] = df[column].str.replace(',', '').astype(float)
    return df

# Example usage:
# Clean commas from 'Close' and 'Adj. Close' columns
Imus_cleaned = clean_commas(Imus, ['Price', 'Open', 'High', 'Low'])
```

```python
def modify_date_column(df, source_column):
    """
    Modifies the date column to the format 'YYYY-MM-DD'.

    Args:
    df (pd.DataFrame): The input DataFrame.
    source_column (str): The name of the column containing the date values.

    Returns:
    pd.DataFrame: DataFrame with the modified date column.
    """
    if source_column in df.columns:
        try:
            # First, convert the column to a datetime object
            df[source_column] = pd.to_datetime(df[source_column], errors='coerce')

            # Then format the datetime column to 'YYYY-MM-DD'
            df[source_column] = df[source_column].dt.strftime('%Y-%m-%d')
        except Exception as e:
            print(f"Error modifying the date column: {e}")
    else:
        print(f"Column '{source_column}' not found in the DataFrame.")
    return df

# Example usage:
GSCI = modify_date_column(GSCI, source_column="Date")

```

```python
def copy_and_rename_column(df, source_column, new_column_name):
    """
    Copies the data from a source column to a new column and renames it.

    Args:
    df (pd.DataFrame): The input DataFrame.
    source_column (str): The name of the column to copy.
    new_column_name (str): The name of the new column.

    Returns:
    pd.DataFrame: DataFrame with the new column added.
    """
    if source_column in df.columns:
        # Copy the source column to the new column
        df[new_column_name] = df[source_column]
    else:
        print(f"Column '{source_column}' not found in the DataFrame.")
    return df

# Example usage:
GSCI = copy_and_rename_column(GSCI, source_column="Close", new_column_name="Adj. Close")
```

```python
def remove_time_from_date(df, date_column):
    """
    Removes the time and timezone portion from the date column, leaving only 'YYYY-MM-DD'.

    Args:
    df (pd.DataFrame): The input DataFrame.
    date_column (str): The name of the column containing the date values.

    Returns:
    pd.DataFrame: DataFrame with the modified date column.
    """
    if date_column in df.columns:
        # Convert to datetime and normalize timezone to UTC, then remove the time
        df[date_column] = pd.to_datetime(df[date_column], utc=True).dt.strftime('%Y-%m-%d')
    else:
        print(f"Column '{date_column}' not found in the DataFrame.")
    return df

# Example usage:
GSCI = remove_time_from_date(GSCI, date_column="Date")
print(GSCI.head())
```
